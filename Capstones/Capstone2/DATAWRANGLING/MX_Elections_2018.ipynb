```python
features_to_keep = [
    'ID_ENTIDAD', 'ENTIDAD', 'ID_DISTRITO_FEDERAL', 'DISTRITO_FEDERAL', 
    'SECCION', 'TIPO_CASILLA', 'UBICACION_CASILLA', 'TOTAL_BOLETAS_SOBRANTES', 
    'TOTAL_PERSONAS_VOTARON', 'PAN', 'PRI',
    'PRD', 'PVEM', 'PT', 'MC', 'PANAL', 'MORENA', 'PES', 'C_PAN_PRD_MC',
    'C_PAN_PRD', 'C_PAN_MC', 'C_PRD_MC', 'C_PRI_PVEM_PANAL', 'C_PRI_PVEM',
    'C_PRI_PANAL', 'C_PVEM_PANAL', 'C_PT_MORENA_PES', 'C_PT_MORENA',
    'C_PT_PES', 'C_MORENA_PES', 'NO_REGISTRADAS', 'NULOS', 'TOTAL_VOTOS_CALCULADO', 'CAND_IND_01', 'CAND_IND_02']
```

```python
df= df[features_to_keep]
```

```python
# Convert columns that should be numeric
columns_to_convert = ['TOTAL_BOLETAS_SOBRANTES', 'TOTAL_PERSONAS_VOTARON', 'PAN', 'PRI', 'PRD', 'PVEM', 
                      'PT', 'MC', 'PANAL', 'MORENA', 'PES', 'C_PAN_PRD_MC', 'C_PAN_PRD', 'C_PAN_MC', 
                      'C_PRD_MC', 'C_PRI_PVEM_PANAL', 'C_PRI_PVEM', 'C_PRI_PANAL', 'C_PVEM_PANAL', 
                      'C_PT_MORENA_PES', 'C_PT_MORENA', 'C_PT_PES', 'C_MORENA_PES', 'NO_REGISTRADAS', 
                      'NULOS', 'TOTAL_VOTOS_CALCULADO', 'CAND_IND_01', 'CAND_IND_02']

for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column], errors='coerce')
```

```python
df.info()
```

```python
candidates_votes = {
    'AMLO': ['PT', 'MORENA', 'PES', 'C_PT_MORENA_PES', 'C_PT_MORENA', 'C_PT_PES', 'C_MORENA_PES'],
    'Meade': ['PRI', 'PVEM', 'PANAL', 'C_PRI_PVEM_PANAL', 'C_PRI_PVEM', 'C_PRI_PANAL', 'C_PVEM_PANAL'],
    'Jaime': ['CAND_IND_02'],  
    'Margarita': ['CAND_IND_01'],  # Actualiza con el nombre correcto si es diferente
    'Anaya': ['PAN', 'PRD', 'MC', 'C_PAN_PRD_MC', 'C_PAN_PRD', 'C_PAN_MC', 'C_PRD_MC']
}

# Verify and sum the votes for each candidate
for candidate, columns in candidates_votes.items():
    existing_columns = [col for col in columns if col in df.columns]
    if not existing_columns:
        print(f"No columns found for candidate {candidate}. Check the column names.")
        continue
    df[candidate] = df[existing_columns].sum(axis=1)

# Drop the original columns related to votes
columns_to_drop = [col for cols in candidates_votes.values() for col in cols if col in df.columns]
df = df.drop(columns=columns_to_drop)
```

```python
# Summarize the votes per candidate
summary = df[['AMLO', 'Meade', 'Jaime', 'Margarita', 'Anaya', 'NULOS']].sum()

# Calculate the total votes across all candidates
total_votes = summary.sum()

# Calculate the percentage of votes per candidate
percentage = (summary / total_votes) * 100

# Combine the summary and percentage into a single DataFrame
summary_df = pd.DataFrame({
    'Total Votes': summary,
    'Percentage': percentage
})

print(summary_df)
```

```python

# Group by state and summarize the votes per candidate
grouped = df.groupby('ENTIDAD')[['AMLO', 'Meade', 'Jaime', 'Margarita', 'Anaya', 'NULOS']].sum()

# Calculate the total votes across all candidates for each state
grouped['Total Votes'] = grouped.sum(axis=1)

# Calculate the percentage of votes per candidate for each state
grouped['AMLO %'] = (grouped['AMLO'] / grouped['Total Votes']) * 100
grouped['Meade %'] = (grouped['Meade'] / grouped['Total Votes']) * 100
grouped['Jaime %'] = (grouped['Jaime'] / grouped['Total Votes']) * 100
grouped['Margarita %'] = (grouped['Margarita'] / grouped['Total Votes']) * 100
grouped['Anaya %'] = (grouped['Anaya'] / grouped['Total Votes']) * 100
grouped['NULOS %'] = (grouped['NULOS'] / grouped['Total Votes']) * 100

# Round the percentage columns to 2 decimal places
grouped = grouped.round({'AMLO %': 2, 'Meade %': 2, 'Jaime %': 2, 'Margarita %': 2, 'Anaya %': 2,'NULOS %': 2 })

# Display the results
grouped.head()
```

```python
grouped = grouped.reset_index()
# Display the DataFrame with improved formatting
pd.set_option('display.colheader_justify', 'center')
grouped.head(32)
```
