```python
import pandas as pd

# Define the file paths
file_2018 = 'grouped_by_state_2018.csv'
file_2024 = 'grouped_by_state_2024.csv'

# Read the CSV files
df_2018 = pd.read_csv(file_2018, encoding='latin1')
df_2024 = pd.read_csv(file_2024, encoding='latin1')

# Display the first few rows of each DataFrame to understand their structure
print("2018 DataFrame:")
df_2018.head(32)
```

```python

print("\n2024 DataFrame:")
df_2024.head()
```

```python
# Inspect the DataFrames
print("Columnas de 2018:")
print(df_2018.columns)

print("\nColumnas de 2024:")
print(df_2024.columns)
```

```python
# Filtrar las columnas relevantes
df_2018 = df_2018[['ENTIDAD', 'AMLO']]
df_2024 = df_2024[['ENTIDAD', 'Claudia','Urban_polling_stations', 'Non_urban_polling_stations']]

# rename 'AMLO' to 'AMLO_2018' to understand is historic data
df_2018.rename(columns={'AMLO': 'AMLO_2018'}, inplace=True)
```

LEFT JOIN OF BOTH dataframes

```python
# Unir los DataFrames en base a la columna 'ENTIDAD'
df_combined = pd.merge(df_2024, df_2018, on='ENTIDAD', how='left')

# Mostrar las primeras filas del DataFrame combinado
df_combined.head(32)
```

```python
economic_data_path = './ECONOMIC_FEATURES/Simplified_PIB_Per_State.csv'
df_economic = pd.read_csv(economic_data_path, encoding='latin1')
df_economic.head()
```

Now the error of not merging correctly is due to capital letters. Lets fix it...

```python
# Convert state names to uppercase
df_2018['ENTIDAD'] = df_2018['ENTIDAD'].str.upper()
df_2024['ENTIDAD'] = df_2024['ENTIDAD'].str.upper()
df_economic['ENTIDAD'] = df_economic['ENTIDAD'].str.upper()
```

```python
# Merge the combined data with economic data
df_with_PIB = pd.merge(df_combined, df_economic, on='ENTIDAD', how='left')
```
